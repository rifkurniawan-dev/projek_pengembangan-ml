```python

nltk.download('punkt_tab')
def cleaningText(text):
    """
    Cleans the input text by removing URLs, mentions, hashtags, and punctuation.

    Args:
        text: The input text string.

    Returns:
        The cleaned text string.
    """

    text = str(text)
    text = re.sub(r'@[A-Za-z0-9]+', '', text)  # Remove mentions
    text = re.sub(r'#', '', text)  # Remove hashtags
    text = re.sub(r'RT[\s]+', '', text)  # Remove retweets
    text = re.sub(r'https?:\/\/\S+', '', text)  # Remove URLs
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    text = emoji.demojize(text, delimiters=("", ""))
    return text
def tokenizingText(text):
    """
    Tokenizes the input text into individual words.

    Args:
        text: The input text string.

    Returns:
        A list of tokens (words) extracted from the text.
    """
    text = word_tokenize(text)
    return text

def filteringText(text):
    """
    Filters out stop words from the input text tokens.

    Args:
        text: A list of tokens (words).

    Returns:
        A list of tokens with stop words removed.
    """
    listStopwords = set(stopwords.words('indonesian'))
    filtered = []
    for txt in text:
        if txt not in listStopwords:
            filtered.append(txt)
    return filtered

def toSentence(list_words):
    """
    Joins a list of words into a single sentence.

    Args:
        list_words: A list of words.

    Returns:
        A single sentence string formed by joining the words.
    """
    sentence = ' '.join(word for word in list_words)
    return sentence

def stemmingText(text):
    """
    Applies stemming to the input text using the Porter Stemmer.

    Args:
        text: The input text string.

    Returns:
        The stemmed text string.
    """
    stemmer = PorterStemmer()
    text = [stemmer.stem(word) for word in text.split()]
    return " ".join(text)

df['text_clean'] = df['content'].apply(cleaningText)
df['text_tokenizingText'] = df['text_clean'].apply(tokenizingText)
df['text_stopword'] = df['text_tokenizingText'].apply(filteringText)
df['content_clean'] = df['text_stopword'].apply(toSentence)
df['stemmingText'] = df['content_clean'].apply(stemmingText)


df.head(1)
```

```python
def wordcloud(data, column):
    """
    Generates and displays a word cloud from a specified column of a pandas DataFrame.

    Args:
        data: The pandas DataFrame containing the text data.
        column: The name of the column containing the text data.
    """
    text = ' '.join(data[column].astype(str).tolist())
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
from collections import Counter

def graph_freq(df, column_name, top_n=20):
    words = ' '.join(df[column_name]).split()
    freq = Counter(words)
    common = freq.most_common(top_n)

    words, counts = zip(*common)

    plt.figure(figsize=(12, 6))
    plt.bar(words, counts)
    plt.xticks(rotation=45)
    plt.title(f"Top {top_n} Most Frequent Words in '{column_name}'")
    plt.xlabel("Words")
    plt.ylabel("Frequency")
    plt.show()
```

```python
def score_label(score):
    """
    Assigns a label based on the given score.

    Args:
        score: The numerical score.

    Returns:
        The corresponding label.
    """
    if score >= 4:
        return 'Positive'
    elif score == 3:
        return 'Neutral'
    else:
        return 'Negative'

def Label(content):
    """
    Assigns a label to the content based on keywords.

    Args:
        content: The text content.

    Returns:
        The assigned label.
    """
    if any(word in content for word in ['mudah', 'membantu', 'bagus', 'aman', 'cepat', 'terbaik', 'puas', 'sangat', 'mantap', 'recomended', 'oke', 'suka']):
        return 'Positive'
    elif any(word in content for word in ['lumayan', 'biasa', 'standar', 'rata', 'cukup']):
        return 'neutral'
    elif any(word in content for word in ['ga', 'gk', 'jelek', 'buruk', 'parah', 'lambat', 'tidak', 'rusak', 'mengecewakan']):
        return 'negative'
    else:
        return 'neutral'

df['score_label'] = df['score'].apply(score_label)
df['label'] = df['content_clean'].apply(Label)
```

```python
#melihat frequency positif review
from collections import Counter

def graph_freq(df, text_col, label_col=None, label_value=None, top_n=20):
    if label_col and label_value is not None:
        df = df[df[label_col] == label_value]

    words = ' '.join(df[text_col]).split()
    freq = Counter(words)
    common = freq.most_common(top_n)

    words, counts = zip(*common)

    plt.figure(figsize=(12, 6))
    plt.bar(words, counts, color='skyblue')
    plt.xticks(rotation=45)
    title = f"Top {top_n} Words in '{text_col}'"
    if label_col and label_value is not None:
        title += f" for {label_col} = '{label_value}'"
    plt.title(title)
    plt.xlabel("Words")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

graph_freq(df, 'content_clean', 'label', 'Positive')
```
